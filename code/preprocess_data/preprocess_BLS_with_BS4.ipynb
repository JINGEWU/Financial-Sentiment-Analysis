```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python

html = urlopen('https://www.bls.gov/bls/news-release/empsit.htm')
bs = BeautifulSoup(html, 'html.parser') 
hyperlink = bs.find_all('a')  
links = []
for h in hyperlink:
    hh = h.get('href')
    links.append(hh)

links   
```

```python
import re
pattern = re.compile(r"\/news.release\/+history+\/empsit_+\d+\.txt")
pattern2 = re.compile(r"\/news.release\/+archives+\/empsit_+\d+\.htm")
result = []
for link in links:
    result1 = re.findall(pattern,link)
    result2 = re.findall(pattern2,link)
    if len(result1)!=0:
        result.append(result1[0])
    if len(result2)!=0:
        result.append(result2[0])
#result = re.findall(pattern,links)
url1 = ['/news.release/archives/empsit_08022019.htm']
url2 = ['/news.release/archives/empsit_04052019.htm']
result.append(url1)
result.append(url2)
print(result[-2:])
```

```python
from requests_html import HTMLSession
session = HTMLSession()
for url in result:
    
#url = 'https://www.bls.gov/news.release/archives/empsit_10022020.htm'
    url_= url[0]
    url_1 = 'https://www.bls.gov' + url_
    r = session.get(url_1)
    file = url_1[-12:-4:1] + '.txt'
    #print(file)
    f=open(file,"w", encoding = 'utf-8')
    f.write(r.html.text)
    f.close()
    print(file,'saved')
```
